```python
def caterpillar_plot_multidat(timings_dict = None, # list [timings_cnn_1024, timings_cnn_4096, timings_mlp_1024, timings_mlp_4096]
                              machine = 'home',
                              n = 1024,
                              save = False,
                              plot_format = 'svg'):
    
    matplotlib.rcParams['text.usetex'] = True
    #matplotlib.rcParams['pdf.fonttype'] = 42
    matplotlib.rcParams['svg.fonttype'] = 'none'
    
    sns.set(style = "white", 
            palette = "muted", 
            color_codes = True,
            font_scale = 2)

    fig, ax = plt.subplots(1, 1, 
                           figsize = (10, 10), 
                           sharex = False, 
                           sharey = False)
    
    my_suptitle = fig.suptitle('Sampler Timings In Seconds', 
                               fontsize = 40)
    
    sns.despine(right = True)
           
    ecdfs = {}
    plot_vals = {'CNN-1024': {}, 
                 'CNN-4096': {},
                 'MLP-1024': {},
                 'MLP-4096': {}} # [0.01, 0.9], [0.01, 0.99], [mean]
    
    for timing_pd_tmp in timings_dict.keys():
        for k in timings_dict[timing_pd_tmp].keys():
            
            if 'RACE' in k or 'LCA' in k:
                if 'MLP' in timing_pd_tmp:
                    continue

            ecdfs[k] = ECDF(timings_dict[timing_pd_tmp][k].values)
            tmp_sorted = sorted(timings_dict[timing_pd_tmp][k])
            _p01 =  tmp_sorted[np.sum(ecdfs[k](tmp_sorted) <= 0.01) - 1]
            _p99 = tmp_sorted[np.sum(ecdfs[k](tmp_sorted) <= 0.99) - 1]
            _p1 = tmp_sorted[np.sum(ecdfs[k](tmp_sorted) <= 0.1) - 1]
            _p9 = tmp_sorted[np.sum(ecdfs[k](tmp_sorted) <= 0.9) - 1]
            _pmean = timings_dict[timing_pd_tmp][k].mean()
            plot_vals[timing_pd_tmp][k] = [[_p01, _p99], [_p1, _p9], _pmean]
        
    x_cnn_1024 = [plot_vals['CNN-1024'][k][2] for k in plot_vals['CNN-1024'].keys()]
    x_cnn_4096 = [plot_vals['CNN-4096'][k][2] for k in plot_vals['CNN-4096'].keys()]

    x_mlp_1024 = [plot_vals['MLP-1024'][k][2] for k in plot_vals['MLP-1024'].keys()]
    x_mlp_4096 = [plot_vals['MLP-4096'][k][2] for k in plot_vals['MLP-4096'].keys()]

    offset_list = [0.4, 0.2, -0.2, -0.4]
    ax.scatter(x_cnn_1024, plot_vals['CNN-1024'].keys(), 
               c = 'black', 
               marker = 's', 
               alpha = .0)
    
    ax.scatter(x_mlp_1024, np.array([i for i in range(len(plot_vals['MLP-1024'].keys()))]) + offset_list[0], 
               c = 'blue',
               marker = 's',
               alpha = .5,
               label = 'MLP-1024')
    
    ax.scatter(x_mlp_4096, np.array([i for i in range(len(plot_vals['MLP-4096'].keys()))]) + offset_list[1], 
               c = 'blue',
               marker = 's',
               alpha = 1.,
               label = 'MLP-4096')
    
    ax.scatter(x_cnn_1024, np.array([i for i in range(len(plot_vals['CNN-1024'].keys()))]) + offset_list[2], 
               c = 'red',
               marker = 's',
               alpha = 0.5,
               label = 'CNN-1024')
    
    ax.scatter(x_cnn_4096, np.array([i for i in range(len(plot_vals['CNN-4096'].keys()))]) + offset_list[3], 
               c = 'red',
               marker = 's',
               alpha = 1,
               label = 'CNN-4096')
    ax.legend()
    ax.set_xlim([0, 1000])
    
#     offset_list = [0.4, 0.15, -0.15, -0.4]
    offset_cnt = 0
    for timing_pd_tmp in timings_dict.keys():
        cnt = 0
        for k in plot_vals[timing_pd_tmp].keys():
            if 'CNN' in timing_pd_tmp:
                tmp_color = 'red'
            else:
                tmp_color = 'blue'
            if '1024' in timing_pd_tmp:
                tmp_alpha = 0.5
                print('passed low alpha')
            else:
                tmp_alpha = 1
            
            ax.plot(plot_vals[timing_pd_tmp][k][0] , 
                    [cnt + offset_list[offset_cnt], cnt + offset_list[offset_cnt]], 
                    c = tmp_color, 
                    zorder = -1, 
                    alpha = tmp_alpha)
            
            ax.plot(plot_vals[timing_pd_tmp][k][1], 
                    [cnt + offset_list[offset_cnt], cnt + offset_list[offset_cnt]], 
                    c = tmp_color, 
                    zorder = -1, 
                    linewidth = 5,
                    alpha = tmp_alpha)



            cnt += 1
        offset_cnt += 1
                                 
    if save == True:
        if machine == 'home':
            fig_dir = "/users/afengler/OneDrive/git_repos/nn_likelihoods/figures/timings/"
            if not os.path.isdir(fig_dir):
                os.mkdir(fig_dir)
        
        if plot_format == 'png':
            plt.savefig(fig_dir + '/' + 'timing_comparison' +  '.png', 
                        dpi = 300, 
                        transparent = False,
                        bbox_inches = 'tight',
                        bbox_extra_artists = [my_suptitle])
        if plot_format == 'svg':
            plt.savefig(fig_dir + '/' + 'timing_comparison' +  '.svg',
                            format = 'svg', 
                            transparent = True,
                            bbox_inches = 'tight',
                            bbox_extra_artists = [my_suptitle],
                            frameon = False)
        
        plt.close()
    
    # Show
        return #plt.show(block = False)
    else:
        return plt.show()
```
